# Function for reading csv file

```python
# Function to read csv file and return data as 2-d Python list
def read_file_housing(filename):
    file_open = open(filename,"r")
    data_array = []
    for line in iter(file_open):
        line_no_newline = line.rstrip()
        line_split = line_no_newline.split(",")
        data_array.append(line_split) 
    file_open.close()
    return data_array
```

###  Create bar chart with zipcode on x-axis and average_price_by_zipcode on y-axis

```python
from statistics import mean 

# Finding Mean by zipcode

HP_95051 = [home_prices_int[_][9] for _ in range(0,len(home_prices_int)) if home_prices_int[_][8] == 95051]
print(HP_95051)

HP_95051_avg = mean(HP_95051)
print(HP_95051_avg)

HP_95014 = [home_prices_int[_][9] for _ in range(0,len(home_prices_int)) if home_prices_int[_][8] == 95014]
print(HP_95014)

HP_95014_avg = mean(HP_95014)
print(HP_95014_avg )

HP_94087 = [home_prices_int[_][9] for _ in range(0,len(home_prices_int)) if home_prices_int[_][8] == 94087]
print(HP_94087)

HP_94087_avg = mean(HP_94087)
print(HP_94087_avg )

HP_94085 = [home_prices_int[_][9] for _ in range(0,len(home_prices_int)) if home_prices_int[_][8] == 94085]
print(HP_94085)

HP_94085_avg = mean(HP_94085)
print(HP_94085_avg)


```

```python
# Adding another column "AvgHomePrices" to home_prices_int

for i in range(0,len(home_prices_int)):
    if home_prices_int[i][8] == 95051 and len(home_prices_int[i])<11:
        home_prices_int[i].append(HP_95051_avg)
    elif home_prices_int[i][8] == 95014 and len(home_prices_int[i])<11:
        home_prices_int[i].append(HP_95014_avg)
    elif home_prices_int[i][8] == 94087 and len(home_prices_int[i])<11:
        home_prices_int[i].append(HP_94087_avg)
    elif home_prices_int[i][8] == 94085 and len(home_prices_int[i])<11:
        home_prices_int[i].append(HP_94085_avg)
        
print(home_prices_int[10:50])
```

```python
# Bucketing home prices
cnt_800_900=0
cnt_900_1000=0
cnt_1000_1100=0
cnt_1100_1200=0
cnt_1200_1300=0
cnt_1300_1400 = 0

for i in range(0,len(HomePriceK)):
    if HomePriceK[i] >800 and HomePriceK[i] <= 900:
        cnt_800_900 = cnt_800_900+1
    elif HomePriceK[i] >900 and HomePriceK[i] <= 1000:
        cnt_900_1000 = cnt_900_1000+1
    elif HomePriceK[i] >1000 and HomePriceK[i] <= 1100:
        cnt_1000_1100 = cnt_1000_1100+1
    elif HomePriceK[i] >1100 and HomePriceK[i] <= 1200:
        cnt_1100_1200 = cnt_1100_1200+1  
    elif HomePriceK[i] >1200 and HomePriceK[i] <= 1300:
        cnt_1200_1300 = cnt_1200_1300+1
    elif HomePriceK[i] >1300 and HomePriceK[i] <= 1400:
        cnt_1300_1400 = cnt_1300_1400+1   
print(cnt_800_900,cnt_900_1000, cnt_1000_1100,cnt_1100_1200,cnt_1200_1300,cnt_1300_1400)        
        
```
